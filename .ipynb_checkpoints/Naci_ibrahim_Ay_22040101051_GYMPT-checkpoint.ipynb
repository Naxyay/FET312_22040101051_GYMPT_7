## Import

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import torch
```

## √áalƒ±≈üma Ortamƒ±

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"√áalƒ±≈üma Ortamƒ±: {device.upper()}")

model = YOLO('yolov8l-world.pt')
```

## G√∂r√ºnt√º ƒ∞≈üleme ve Tahmin

```python
image_path = "deneme.jpg"

results = model.predict(source=image_path, save=False, conf=0.05, iou=0.5, device=0)
```

## √áƒ±ktƒ±

```python
if len(results) > 0:
    result = results[0]
    
    for box in result.boxes:
        class_id = int(box.cls[0])
        class_name = result.names[class_id]
        confidence = float(box.conf[0])
        print(f"üìç {class_name}: %{confidence*100:.2f}")

    res_plotted = result.plot()
    res_rgb = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(12, 12))
    plt.imshow(res_rgb)
    plt.axis('off')
    plt.title("Tespit Sonucu")
    plt.show()

else:
    print("Belirtilen nesneler bulunamadƒ±.")
```
